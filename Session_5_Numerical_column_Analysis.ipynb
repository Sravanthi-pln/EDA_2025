```python
#u-1*sigma to u+1*sigma the data coverage 68%
#step-1: calculate wage mean =u
#step-2: calculate std of wage = sigma
#step-3: lowerbound =u-1*sigma
#step-4: upperbound=u+1*sigma
#step-5: cond1=Wage_data>lower bound
#step-6: cond2=wage_data< upper bound
#step-7: con=cond1&cond2
#step-8: len(visa_data[cond]
```

```python
wage_df=visa_df['prevailing_wage']

def Emperical_rule(k,n):
    wage_mean=np.mean(wage_df)
    wage_std=np.std(wage_df)
    
    lb=wage_mean-k*wage_std
    ub=wage_mean+k*wage_std
    
    cond1=wage_data>=lb
    cond2=wage_data<=ub
    con=cond1&cond2
    print(len(visa_df[con])==n*len(visa_df)/100)

Emperical_rule(3,99.7)
```
