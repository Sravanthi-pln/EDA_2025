```python
lables = visa_df['continent'].unique()

l=[]

for i in lables:
    con=visa_df['continent'] == i
    l.append(len(visa_df[con]))

l
```

```python
cols=['continent', 'count']

continent_df=pd.DataFrame(zip(lables,l),columns=cols)
continent_df
```

**pie chat**

```python
visa_df['continent'].value_counts(normalize=True)
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,
        labels=keys,
        explode=[0.1 for i in range(6)],
        autopct="%0.2f%%",
        startangle=180,
       radius=2)
plt.show()
```
