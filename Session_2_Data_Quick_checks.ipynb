**Task -2 : print the columns if it is object from visa_df**

```python
visa_df.dtypes.to_dict()
for i,j in visa_df.dtypes.to_dict().items():
    if j=='object':
        print(i)
```

```python
obj=[]
num=[]
data_type=visa_df.dtypes.to_dict().items()
for i,j in data_type:
    if j=='object':
        obj.append(i)
    else:
        num.append(i)

print('object list:', obj)
print('object list:', num)
```

**List Comprehension**

```python
data_type=visa_df.dtypes.to_dict().items()
obj=[i for i,j in data_type if j=='object']
num=[i for i,j in data_type if j!='object']

print('object list:', obj)
print('object list:', num)
```
